2. Agora seguiremos para o exercício proposto, onde vou carregar três arquivos neste código e concatená-los. Baixei do site do INMET três arquivos .csv com dados metereológios de três capitais: Curitiba - PR, Salvador - BA e Vitória - ES. O código abaixo permite selecionar qualquer número de arquivos, de qualquer formato, para carregamento dentro do programa.

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

O bloco abaixo concatena, entre os arquivos carregados, todos que apresentam a extensão .csv, em um novo .csv.

```python
#combinar todos os arquivos da lista
combined_csv = pd.concat([pd.read_csv(f, sep=';') for f in uploaded.keys() if f.endswith('.csv')])
#exportar para csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

Abaixo, podemos realizar o download do novo .csv, resultante da concatenação.
