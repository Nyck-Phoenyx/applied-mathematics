```python
# Resfriamento de Newton
# Autor: Nyckolas Phoenyx
# Data: 05/09/2025

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import math
```

```python
# ----------------------------
# 1. Modelo matemático
# ----------------------------
# Equação do resfriamento de Newton:
# dT/dt = -k * (T - T_env)

def newton_cooling(T, t, k, T_env):
    return -k * (T - T_env)
```

```python
# ----------------------------
# 2. Parâmetros do problema
# ----------------------------
T0 = float(input("Temperatura Inicial do Objeto (C°): "))
T_env = float(input("Temperatura Ambiente (C°): "))
k = float(input("Constante de resfriamento: "))
min = float(input("Tempo de simulação (em minutos): "))
t = np.linspace(0, min, 200)  # Tempo (minutos)
```

```python
# ----------------------------
# 3. Solução numérica
# ----------------------------
T = odeint(newton_cooling, T0, t, args=(k, T_env))
```

```python
# ----------------------------
# 4. Visualização
# ----------------------------
plt.figure(figsize=(8,5))
plt.plot(t, T, label="Temperatura do objeto", color="blue")
plt.axhline(T_env, color="red", linestyle="--", label="Temperatura ambiente")
plt.xlabel("Tempo (min)")
plt.ylabel("Temperatura (°C)")
plt.title("Resfriamento de Newton")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ----------------------------
# 5. Determinar o tempo necessário para uma dada temperatura
# ----------------------------

T1 = float(input("Temperatura Final do Objeto: "))
dt = -(1/k) * math.log((T1 - T_env) / (T0 - T_env))
print(f"{dt:.2f} minutos")

```
